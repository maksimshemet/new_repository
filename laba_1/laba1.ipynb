Для кожної із адміністративних одиниць України завантажити тестові структуровані файли, що містять значення VHI-індексу. Ця процедура має бути автоматизована, параметром процедури має бути індекс (номер) області. При зберіганні файлу до його імені потрібно додати дату та час завантаження

```python
from urllib.request import urlopen
import pandas as pd
import collections
from datetime import datetime

now = datetime.now()
oblasti = (
    "1: Cherkasy",
    "2: Chernihiv",
    "3: Chernivtsi",
    "4: Crimea",
    "5: Dnipropetrovs'k",
    "6: Donets'k",
    "7: Ivano-Frankivs'k",
    "8: Kharkiv",
    "9: Kherson",
    "10: Khmel'nyts'kyy",
    "11: Kiev",
    "12: Kiev City",
    "13: Kirovohrad",
    "14: Luhans'k",
    "15: L'viv",
    "16: Mykolayiv",
    "17: Odessa",
    "18: Poltava",
    "19: Rivne",
    "20: Sevastopol'",
    "21: Sumy",
    "22: Ternopil",
    "23: Transcarpathia",
    "24: Vinnytsya",
    "25: Volyn",
    "26: Zaporizhzhya",
    "27: Zhytomyr"
                )
for i in oblasti:
    print(i)
id = input("Введите ID области:")
url = "https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_provinceData.php?country=UKR&provinceID={}&year1=1981&year2=2017&type=Mean".format(id)
d ='%s-%s-%s' % (now.month, now.day, now.year)
t = '%s:%s:%s' % (now.hour, now.minute, now.second)
destination = t+"_"+d+"_"+str(oblasti[int(id)-1])+".csv"
name = destination.replace(" ","_")
print(name)
vhi_url = urlopen(url)
out = open(name, "wb")
out.write(vhi_url.read())
out.close()
print ("VHI is downloaded...")




```

```python
#вивід таблиці
df = pd.read_csv(name , index_col=False, header=1, skipfooter=1, engine='python',
                 names=['year', 'week', 'SMN', 'SMT', 'VCI', 'TCI', 'VHI'], delimiter='\,\s+|\s+|\,')
df
```

```python
#Ряд VHI для області за рік, пошук екстремумів (min та max)
year = input("Введите год: ")
df2 = df[(df['year']==int(year))]
print("Максимальный VHI: "+str(max(df2['VHI'])))
print("Минимальный VHI: "+str(min(df2['VHI'])))

```
